```python
# 1. Clean and standardization of all dates to 'YYYY-MM-DD' format



dateFormats = ['%Y-%m-%d', '%d-%m-%Y', '%m-%d-%Y',
               '%Y/%m/%d', '%d/%m/%Y', '%m/%d/%Y',
               '%Y.%m.%d', '%d.%m.%Y', '%m.%d.%Y'
               ] 
df12['original_order_date'] = df12['order_date']

def parseDate(x):
    x = str(x).strip()
    for date in dateFormats:
        try:
            return pd.to_datetime(x, format=date)
        except:
            continue
    return pd.NaT

df12['order_date'] = df12['order_date'].apply(parseDate)
df12['is_valid_date'] = df12['order_date'].notna()
```

```python
# 2. Cleaning column to contain only numeric values in Indian Rupees.


def parsePrice(x):
    x = str(x).replace(",","").strip()
    x = ''.join(c for c in x if c.isdigit() or c=='.')
    try:
        return float(x)
    except:
        return pd.NA

df12['before_parse_price'] = df12['original_price_inr']
df12['original_price_inr'] = df12['original_price_inr'].apply(parsePrice)
```

```python
# 3.Standardization of all ratings to numeric scale 1.0-5.0


def parseCustomerRating(x):
    x = str(x).replace("/5","").strip()
    i=0
    x1=""
    for c in x:
        x1+=c
        i+=1
        if i==3:
            break
    return float(x1)

df12['original_customer_rating'] = df12['customer_rating']
df12['customer_rating'] = df12['customer_rating'].apply(parseCustomerRating)
#Filling null values with overall product rating from dataset
df12['customer_rating'] = df12['customer_rating'].fillna(df12['product_rating'])
```

```python
# 4.Standardization of  all city names  to its original


df12['customer_city'] = df12['customer_city'].replace(
    ["DELHI", "Delhi NCR", "delhi", "New Delhi" ],
      "Delhi"
)

df12['customer_city'] = df12['customer_city'].replace(
    ["Kolkata ", "KOLKATA", "Calcutta", "kolkata" ],
      "Kolkata"
)

df12['customer_city'] = df12['customer_city'].replace(
    ["Madras", "Chennai ", "chenai", "CHENNAI" ],
      "Chennai"
)

df12['customer_city'] = df12['customer_city'].replace(
    ["Mumbai ", "MUMBAI", "mumba", "Bombay" ],
      "Mumbai"
)

df12['customer_city'] = df12['customer_city'].replace(
    ["Bengaluru", "BANGALORE", "Bengalore", "Banglore" ],
      "Bangalore"
)
```

```python
# 5. Converting all boolean columns to consistent True/False format.



df12['is_prime_member'] = df12['is_prime_member'].replace(
    ["0", "No","False ","FALSE",], "False"
)
df12['is_prime_member'] = df12['is_prime_member'].replace(
    ["1", "Yes","True ","TRUE"], "True"
)
df12['is_prime_eligible'] = df12['is_prime_eligible'].replace(
    ["0", "No","False ","FALSE",], "False"
)
df12['is_prime_eligible'] = df12['is_prime_eligible'].replace(
    ["1", "Yes","True ","TRUE"], "True"
)
df12['is_festival_sale'] = df12['is_festival_sale'].replace(
    ["0", "No","False ","FALSE",], "False"
)
df12['is_festival_sale'] = df12['is_festival_sale'].replace(
    ["1", "Yes","True ","TRUE"], "True"
)
```

```python
# 6. '. Standardization of category names across the dataset and ensure consistent naming conventions.



df12['original_delivery_days'] = df12['delivery_days']
day=['0','1','2','3','4','5','6','7']
def delivery(row):
    days = str(row['delivery_days'])
    type = row['delivery_type']
    prime_user = row['is_prime_member']
    if days in day:
       return int(days)
    elif days == '-1':
        if type == 'Same Day' or 'Express':
            return 0
        elif type == 'Standard' and prime_user is True:
            return 1
        else:
            return 2
    elif days == '1-2 days':
        if type == 'Same Day' or 'Express':
            return 1
        elif type == 'Standard' and prime_user is True:
            return 2
        else:
            return 3
    elif days == 'Express':
        if type == 'Same Day' or 'Express':
            return 0
        elif type == 'Standard' and prime_user is True:
            return 1
        else:
            return 2
    elif days == 'Same Day':
        if type == 'Same Day' or 'Express':
            return 0
        elif type == 'Standard' and prime_user is True:
            return 1
        else:
            return 2
    elif days == '15':
        if type == 'Same Day' or 'Express':
            return 1
        elif type == 'Standard' and prime_user is True:
            return 3
        else:
            return 5

df12['delivery_days'] = df12.apply(delivery, axis=1)
```

```python
# 8. Identifing and handling duplicate transactions where the same customer, product, date, and amount appear multiple times

df12 = df12.drop_duplicates(subset=['customer_id','product_name','order_date'], keep='first')
df12 = df12.reset_index(drop=True)
df12.info()
```

```python
# 9 . Outlier Detection and Removal using IQR method
# df = df.copy()

#  Compute expected price (per unit)
df12['expected_price'] = df12['final_amount_inr'] / df12['quantity']

#  IQR bounds for outlier detection
Q1 = df12['original_price_inr'].quantile(0.25)
Q3 = df12['original_price_inr'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Get outliers first (don’t drop yet)
outliers = df12[(df12['original_price_inr'] < lower_bound) | (df12['original_price_inr'] > upper_bound)].copy()

#  Function to correct decimal errors
def correct_decimal(row):
    orig = row['original_price_inr']
    exp = row['expected_price']
    
    if pd.isna(exp) or exp == 0:  # avoid invalid division
        return orig
    
    ratio = round(orig / exp, 2)
    if ratio == 10:
        return orig / 10
    elif ratio == 100:
        return orig / 100
    elif ratio == 1000:
        return orig / 1000
    else:
        return orig

#  Correct only the outliers
outliers['corrected_price'] = outliers.apply(correct_decimal, axis=1)

#  Update corrected prices back into df
df12.loc[outliers.index, 'original_price_inr'] = outliers['corrected_price']

#  Re-check IQR after correction and remove remaining true outliers
Q1 = df12['original_price_inr'].quantile(0.25)
Q3 = df12['original_price_inr'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df12 = df12[(df12['original_price_inr'] >= lower_bound) & (df12['original_price_inr'] <= upper_bound)].copy()

#  Clean final DataFrame
df12.drop(['expected_price'], axis=1, inplace=True)
df12.reset_index(drop=True, inplace=True)

df12.info()
```
